# Basics of Matplotlib

```python
#Scatter plot
import numpy as np
import matplotlib.pyplot as plt

x_data = np.random.random(100) * 100
y_data = np.random.random(100) * 100


plt.scatter(x_data, y_data, c ='red',marker="*",s=100)
plt.show()
```

## Scatter plot

```python
#Scatter plot
import numpy as np
import matplotlib.pyplot as plt


plt.scatter(x=[1,2,3,4], y=[1,2,3,4], c ='red',marker="*",s=100)
plt.show()
```

## Line plot

```python
import numpy as np
import matplotlib.pyplot as plt
years = [2006 + x for x in range(16)]
print(years, len(years))
weights = np.random.random(16)
plt.plot(years,weights)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
years = [2006 + x for x in range(16)]
print(years, len(years))
weights = np.random.random(16)
plt.plot(years,weights,lw=3,linestyle='--')
plt.scatter(years,weights,c='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
years = [2006 + x for x in range(16)]
print(years, len(years))
weights = np.random.random(16)
plt.plot(years,weights,'g--',lw=3)
plt.scatter(years,weights,c='red')
plt.show()
```

## Bar chart

```python
import numpy as np
import matplotlib.pyplot as plt
x = ['C++','C#','Python','Java','Go']
y = [20,50,140,12,34]
plt.xlabel('Programming Languages')
plt.ylabel('Number of users in million')
plt.bar(x,y, color= 'violet',align='edge',width=0.5,edgecolor='black',lw=3)
plt.show()
```

## Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
ages = np.random.normal(20,1.5,1000)
#print(ages)
plt.hist(ages,
         bins=50)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
ages = np.random.normal(20,1.5,1000)
#print(ages)
plt.hist(ages,
         bins=[ages.min(), 18,21,ages.max()])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
ages = np.random.normal(20,1.5,1000)
#print(ages)
plt.hist(ages,
         bins=20,
         cumulative=True)
plt.show()
```

## Plotting categorical data

```python
import numpy as np
import matplotlib.pyplot as plt
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

## Pie Chart

```python
import numpy as np
import matplotlib.pyplot as plt

langs = ['Python','C++','Java','C#','Go']
votes = [50,24,14,6,17]

plt.pie(votes,labels=langs)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

langs = ['Python','C++','Java','C#','Go']
votes = [50,24,14,6,17]
explodes = [0,0,0,0.2,0]

plt.pie(votes,labels=langs,explode=explodes)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

langs = ['Python','C++','Java','C#','Go']
votes = [50,24,14,6,17]
explodes = [0,0,0,0.2,0]

plt.pie(votes,labels=langs,explode=explodes,autopct="%.2f%%",pctdistance=1.3)
plt.show()
```

## Box plot

```python
heights = np.random.normal(172,8,300)
#print(heights)
plt.boxplot(heights)
plt.show()
```

## linspace

```python
import numpy as np
values = np.linspace(start=0,stop=9,num=10)
print(values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

first = np.linspace(0,10,25)
second = np.linspace(10,100,25)
third = np.linspace(200,210,25)
fourth = np.linspace(210,230,25)


data = np.concatenate((first,second,third,fourth))
plt.boxplot(data)
plt.show()
```
